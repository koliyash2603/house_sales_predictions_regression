```python
outlier_columns=['MSSubClass','LotFrontage', 'LotArea','OverallCond', 'YearBuilt','MasVnrArea','BsmtFinSF1','BsmtUnfSF','TotalBsmtSF','1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath','BsmtHalfBath', 'BedroomAbvGr', 'KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea','WoodDeckSF', 'OpenPorchSF','EnclosedPorch', '3SsnPorch',
       'ScreenPorch','PoolArea', 'MiscVal']
```

```python
def outlier_remove(df, columns):
    df_copy = df.copy()

    for column in columns:
        # Calculate the first and third quartiles (Q1 and Q3)
        Q1 = df_copy[column].quantile(0.25)
        Q3 = df_copy[column].quantile(0.75)
	
        # Calculate the interquartile range (IQR)
        IQR = Q3 - Q1

        # Define the lower and upper bounds to identify outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Cap the outliers by replacing them with the nearest valid value within the bounds
        df_copy[column] = np.clip(df_copy[column], lower_bound, upper_bound)

    return df_copy
```

```python
df1=outlier_remove(con,outlier_columns)
df1
```

# Preprocessing

# Preprocessing : Pipeline Building

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
# Num-Pipe
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scale',StandardScaler())])
# cat_Pipe:
cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('encode',OrdinalEncoder())])
# Column_transformer
pre=ColumnTransformer([('num',num_pipe,con2),('cat',cat_pipe,cat2)])
```

```python
x3=pre.fit_transform(x2)
x3
```

```python
pre_cols=pre.get_feature_names_out()
```

```python
x3=pd.DataFrame(x3,columns=pre_cols)
```

```python
x3
```

# Train-Test-Split

```python
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV

x_train,x_test,y_train,y_test=train_test_split(x3,y,random_state=42,test_size=0.2)
```

# Model-Selection

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge

lr=LinearRegression()
la=Lasso()
ri=Ridge()
```

```python
model=[lr,la,ri]
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
for i in model:
    i.fit(x_train,y_train)
    y_pred_tr=i.predict(x_train)
    y_pred_ts=i.predict(x_test)
    print('model:',i)
    mse_tr=mean_squared_error(y_pred_tr,y_train)
    print('mse_tr:',round(mse_tr,2))
    mae_tr=mean_absolute_error(y_pred_tr,y_train)
    print('mae_tr:',round(mae_tr,2))
    r2_tr=r2_score(y_pred_tr,y_train)
    print('r2_tr:',round(r2_tr,2))
    print('__'*10)
    mse_ts=mean_squared_error(y_pred_ts,y_test)
    print('mse_ts:',round(mse_ts,2))
    mae_ts=mean_absolute_error(y_pred_ts,y_test)
    print('mae_ts:',round(mae_ts,2))
    r2_ts=r2_score(y_pred_ts,y_test)
    print('r2_ts:',round(r2_ts,2))
    print('__'*62)
```
